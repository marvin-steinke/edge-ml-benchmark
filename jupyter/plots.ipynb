```python
import pandas as pd
import matplotlib.colors as mcolors
import matplotlib.pyplot as plt

batches = [8, 16, 32, 48, 64]
colors = list(mcolors.TABLEAU_COLORS.values())
dfs = [pd.read_csv(f'../plotdata/{batch}.csv') for batch in batches]
```

```python
fig, axs = plt.subplots()
for i, df in enumerate(dfs):
    df.plot(
        x = 'frequency',
        xlabel = 'Frequency in MHz',
        y = 'mean_consumption',
        ylabel = 'Mean Power Consumption in W',
        legend = True,
        ax = axs,
        color = colors[i],
        title = 'PyTorch CIFAR-10 on RPi 3b+, 1 Round with Diff. Batchsizes',
    )
axs.legend(batches)
fig.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots()
for i, df in enumerate(dfs):
    df.plot(
        x = 'frequency',
        xlabel = 'Frequency in MHz',
        y = 'runtime',
        ylabel = 'Runtime in s',
        legend = True,
        ax = axs,
        color = colors[i],
        title = 'PyTorch CIFAR-10 on RPi 3b+, 1 Round with Diff. Batchsizes',
    )
axs.legend(batches)
fig.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots()
for i, df in enumerate(dfs):
    df['mean_energy'] = df['runtime'] * df['mean_consumption']
    df.plot(
        x = 'frequency',
        xlabel = 'Frequency in MHz',
        y = 'mean_energy',
        ylabel = 'Energy in Ws',
        legend = True,
        ax = axs,
        color = colors[i],
        title = 'PyTorch CIFAR-10 on RPi 3b+, 1 Round with Diff. Batchsizes',
    )
axs.legend(batches)
fig.tight_layout()
plt.show()
```
